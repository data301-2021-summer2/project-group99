```python
def df_column_uniquify(df):
    df_columns = df.columns
    new_columns = []
    for item in df_columns:
        counter = 0
        newitem = item
        while newitem in new_columns:
            counter += 1
            newitem = "{}_{}".format(item, counter)
        new_columns.append(newitem)
    df.columns = new_columns
    return df
df = df_column_uniquify(df)
df.columns
```

## Analysis

Through my analysis of the data referring to mlb predictions I learned many things. Prior to this, I knew mlb predictions took many factors into account, but never knew how exactly they played out. At the beginning, I was most interested in how the starting pitcher affected the predicted outcome of the game. After working with the data, checking how accurately the predictions were seemed to appeal to me more.

After doing the initial cleaning of the data, I first checked how many of the games played out in the same outcome as the system predicted. The system has a model that predicts the probability based on the team, as well as an updated model that takes into account who is the starting pitcher that day. What I found was that out of 1347 games thus far in the season, accounting for the starting pitcher predicted one less outcome than not, an unexpected result. These results were correct around 57.5 percent of the time. This led me to be curious as to how the team with the better pitcher performed, regardless of the rest of the team. I found that if you only look at the rating of the pitcher, you’d be about 61 percent right, a large improvement over what should be a better system. 

Next, after noticing that the system consistently updates its ratings based on season performance thus far, I was led to think that as the season continues, the system would get better at predicting outcomes. I plotted how many games were predicted correctly based on date and found there was no noticeable increase in ability to predict games.

Another item I wanted to check is the average advantage of the favourite to win the game. This told me the average odds for the favoured team to win are 57.7 percent, incredibly close to the 57.5 percent of correct outcomes, meaning these odds are overall very accurate in the long run. I then ran a check on the heavy favourites (those with over 62.5 percent chance of winning) and found that with average odds of 66.2 percent, 64.6 percent of them actually won, again confirming the accuracy of these predictions.

I was also curious about how a team's rating changes throughout the year. I plotted the team rating with the date of the Toronto Blue Jays and noticed the ratings change quite a bit game to game, meaning if a team is on a winning streak their rating would increase dramatically. This would also have a great effect on their odds of winning.
