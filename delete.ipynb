```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("test.csv")
```

```python
plt.figure(figsize=(20,5))
```

```python
sns.despine()
```

```python
plt.figure(figsize=(20,5))
sns.despine()
plt.xlabel('Species')
plt.ylabel('Bill Length')
plt.title('Penguin Bill Length by species')
vp = sns.violinplot(data=test, x = 'Species', y='Bill Length', hue = 'Sex', split = True)
plt.legend(loc='upper left')
```
